## Crear tablas (36 pts.)

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "facciones" (6 pts)

```python
%%sql

CREATE TABLE facciones (
    id INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    descripcion VARCHAR(45) NOT NULL,
    PRIMARY KEY (id)
);
```

### Tabla "trabajos" (6 pts)

```python
%%sql

CREATE TABLE trabajos (
    id INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    descripcion VARCHAR(45) NOT NULL,
    sueldo INTEGER NOT NULL,
    PRIMARY KEY (id)
);
```

### Tabla "personas" (6 pts)

```python
%%sql

CREATE TABLE personas (
    id INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    apellidos VARCHAR(45) NOT NULL,
    fecha_nacimiento DATE NOT NULL,
    infectado BOOL NOT NULL,
    conyugue INTEGER,
    PRIMARY KEY (id),
    FOREIGN KEY (conyugue) REFERENCES personas(id)
);
```

### Tabla "cuentas_bancarias" (6 pts)

```python
%%sql

CREATE TABLE cuentas_bancarias (
    id INTEGER GENERATED BY DEFAULT AS IDENTITY,
    tipo_de_cuenta VARCHAR(45) NOT NULL,
    banco VARCHAR(45) NOT NULL,
    saldo INTEGER NOT NULL,
    id_persona INTEGER,
    PRIMARY KEY (id),
    FOREIGN KEY (id_persona) REFERENCES personas(id)
);
```

### Tabla "persona_pertenece_faccion" (6 pts)

```python
%%sql

CREATE TABLE persona_pertenece_faccion (
    id_persona INTEGER GENERATED BY DEFAULT AS IDENTITY,
    id_faccion INTEGER GENERATED BY DEFAULT AS IDENTITY,
    PRIMARY KEY (id_persona,id_faccion),
    FOREIGN KEY (id_persona) REFERENCES personas(id),
    FOREIGN KEY (id_faccion) REFERENCES facciones(id)
);
```

### Tabla "persona_tiene_trabajo" (6 pts)

```python
%%sql

CREATE TABLE persona_tiene_trabajo (
    id_persona INTEGER GENERATED BY DEFAULT AS IDENTITY,
    id_trabajo INTEGER GENERATED BY DEFAULT AS IDENTITY,
    estado BOOL NOT NULL,
    ultima_vez_realizado TIMESTAMP NOT NULL,
    PRIMARY KEY (id_persona,id_trabajo),
    FOREIGN KEY (id_persona) REFERENCES personas(id),
    FOREIGN KEY (id_trabajo) REFERENCES trabajos(id)
);
```

### Consulta 1 (6 pts)
**ID, nombre y primer apellido de los habitantes de FEDRA**

Ejemplo:

|ID |  Nombre      |  Apellido  |

| 12 |Joel | Miller |

**Solución:** _Realizamos un inner join entre las tablas "personas", "persona_pertenece_faccion" y "facciones", del cual seleccionamos el id, nombre, y apellido, recortando apellidos hasta donde encuentre un espacio en blanco. Finalmente ponemos como condicion que estas personas pertenezcan a FEDRA._


```python
%%sql

SELECT 
	personas.id AS "ID", 
	personas.nombre AS "Nombre", 
	SUBSTRING(personas.apellidos,0,strpos(personas.apellidos,' ')) "Apellido"
FROM 
	personas 
	INNER JOIN persona_pertenece_faccion
	ON personas.id = persona_pertenece_faccion.id_persona
	INNER JOIN facciones
	ON facciones.id = persona_pertenece_faccion.id_faccion
WHERE 
	facciones.nombre = 'FEDRA'
ORDER BY 
	personas.id;
```

### Consulta 4 (6 pts)
**Nombre de la Facción y la cantidad de trabajadores activos hasta fecha X.**

Ejemplo:

| Facción     | Cantidad de trabajadores |
|----------------|--------|
| FEDRA | 1273 |
| Luciérnagas | 121 |

**Solución:** _Realizamos un inner join entre personas, persona_pertenece_faccion, facciones y persona_tiene_trabajo. Pedimos la faccion y la cantidad de trabajadores, la cual la calculamos haciendo un count a persona_pertenece_faccion.id_persona. Ponemos como condición que la persona haya nacido antes de la fecha pedida y que esta esté actualmente trabajando._

```python
%%sql


SELECT 
	facciones.nombre AS "Facción", 
	COUNT(persona_pertenece_faccion.id_persona) AS "Cantidad de trabajadores"
FROM
	personas INNER JOIN persona_pertenece_faccion
	ON personas.id = persona_pertenece_faccion.id_persona
	INNER JOIN facciones
	ON persona_pertenece_faccion.id_faccion = facciones.id
	INNER JOIN persona_tiene_trabajo
	ON personas.id = persona_tiene_trabajo.id_persona
WHERE personas.fecha_nacimiento < '2015-10-06' AND persona_tiene_trabajo.estado = TRUE
GROUP BY facciones.nombre;
```

### Consulta 5 (6 pts)
**Nombre de la Facción y su cantidad de personas infectadas.**

Ejemplo:

|  Facción  | Personas infectadas | 
|--------|-------|
| FEDRA | 251 |
| Luciérnagas | 17 |
| Carroñeros  | 12 |
| ...  | ... |

**Solución:** _Primeramente explicaremos las inner join de las consultas:

Generamos una tabla seleccionando los atributos id_persona e id_faccion desde un INNER JOIN entre las tablas "personas" y "persona_pertenece_facción", considerando sólo aquellas personas que estén infectadas a la cual llamamos sub_union

Luego, desde un INNER JOIN entre "sub_union" y "facciones" generamos otra tabla seleccionando el nombre (de la facción), la cantidad de personas infectadas (COUNT id_persona), agrupamos la tabla anterior por el nombre de la faccion y la ordenamos de manera descendiente según cantidad de personas infectadas._

```python
%%sql

SELECT
    nombre AS "Facción", COUNT(id_persona) AS "Personas infectadas"
FROM(
    SELECT 
        id_persona, id_faccion
    FROM
        personas INNER JOIN persona_pertenece_faccion on id = id_persona
    WHERE
        infectado = True
    )
    AS sub_union INNER JOIN facciones on id_faccion = id

GROUP BY
    nombre 

ORDER BY
    COUNT(id_persona) DESC;
```

### Consulta 6 (6 pts)
**Nombre de la Facción y su trabajo más realizado.**

Ejemplo:

|  Facción  | Trabajo | 
|--------|-------|
| FEDRA | Limpiador de cloacas |
| Luciérnagas | Médico |
| Carroñeros  | Explorador |
| ...  | ... |

**Solución:** _Seleccionamos trabajo, cantidad_trabajadores y faccion desde un inner join de personas, persona_tiene_trabajo, trabajos, persona_pertenece_faccion y facciones (gigante), esta tabla la llamamos nueva_tabla y nos muestra la cantidad de trabajadores de cada oficio por facción. Luego de esta tabla seleccionamos trabajo, cantidad_trabajadores y faccion donde la cantidad de trabajadores sea maxima, para esto utilizamos un ANY dentro del cual se encuentra una tabla que contiene las facciones y su respectiva mayor cantidad de trabajadores. Finalemente (T-T) seleccionamos faccion y trabajo, como algunos trabajos empatan en cantidad maxima de trabajadores para cada faccion, escogemos uno utilizando la funcion MAX en trabajo._

```python
%%sql

SELECT  tablita.faccion AS "Facción", 
		MAX(tablita.trabajo) AS "Trabajo"
FROM (

    SELECT
        nueva_tabla.faccion AS faccion,
        nueva_tabla.cantidad_trabajadores AS cantidad_trabajadores,
        nueva_tabla.trabajo AS trabajo
    FROM(
        SELECT
            trabajos.nombre AS trabajo,
            COUNT(persona_tiene_trabajo.id_persona) AS cantidad_trabajadores,
            facciones.nombre AS faccion
        FROM
            personas INNER JOIN persona_tiene_trabajo
            ON personas.id = persona_tiene_trabajo.id_persona
            INNER JOIN trabajos
            ON persona_tiene_trabajo.id_trabajo = trabajos.id
            INNER JOIN persona_pertenece_faccion
            ON personas.id= persona_pertenece_faccion.id_persona
            INNER JOIN facciones
            ON persona_pertenece_faccion.id_faccion = facciones.id
        GROUP BY facciones.nombre, trabajos.nombre
        ) AS nueva_tabla
    WHERE (nueva_tabla.faccion,nueva_tabla.cantidad_trabajadores) = ANY(
        SELECT
            nueva_tabla1.faccion AS "Faccion",
            MAX(nueva_tabla1.cantidad_trabajadores) AS "Cantidad Trabajadores"
        FROM(
            SELECT
                trabajos.nombre AS trabajo,
                COUNT(persona_tiene_trabajo.id_persona) AS cantidad_trabajadores,
                facciones.nombre AS faccion
            FROM
                personas INNER JOIN persona_tiene_trabajo
                ON personas.id = persona_tiene_trabajo.id_persona
                INNER JOIN trabajos
                ON persona_tiene_trabajo.id_trabajo = trabajos.id
                INNER JOIN persona_pertenece_faccion
                ON personas.id= persona_pertenece_faccion.id_persona
                INNER JOIN facciones
                ON persona_pertenece_faccion.id_faccion = facciones.id
            GROUP BY facciones.nombre, trabajos.nombre
            ) AS nueva_tabla1
            GROUP BY nueva_tabla1.faccion
            )) AS tablita
GROUP BY tablita.faccion;
```

### Consulta 7 (8 pts)
**Nombre de la Facción y su Saldo total** (Es decir, la suma total de los saldos de cada persona perteneciente a la facción.)

Ejemplo:

|Facción | Saldo |
|--------|-|
| FEDRA | 523000 |
| Luciérnagas | 130023 |
| Carroñeros  | 17432 |

**Solución:** _Primeramente explicaremos los inner joins interiores:

Generamos una tabla a partir de un INNER JOIN entre las tablas "cuentas_bancarias" y "persona_pertenece_faccion" conisderando los id de personas, Agrupamos considerando el id_faccion, y seleccinoamos id_faccion y la suma de los saldos por cada faccion, la tabla anterior la llamamos sub_union.

Realizamos un INNER JOIN entre las tablas "sub_union" y "facciones" considerando los id de las facciones, seleccionamos el nombre (de la faccion) y la suma de los saldo por faccion ("Saldo")
además, para mayor orden ordenamos los saldos de manera descendiente._

```python
%%sql

SELECT
    nombre AS "Facción", "Saldo"
FROM(
    SELECT
        id_faccion, SUM(saldo) AS "Saldo"
    FROM
        cuentas_bancarias INNER JOIN persona_pertenece_faccion ON cuentas_bancarias.id_persona = persona_pertenece_faccion.id_persona
    GROUP BY
        id_faccion
    ) AS sub_union

    INNER JOIN facciones on id_faccion = id

ORDER BY
    "Saldo" DESC;
```

**Solución:** _Realizamos un inner join entre personas y cuentas_bancarias, del cual seleccionamos el banco, la cantidad de cuentas bancarias utilizando una función count en cuentas_bancarias.id_persona, y el saldo total de este utilizando una función SUM en cuentas_bancarias.saldo, todo esto lo agrupamos por banco y lo ordenamos por saldo._

```python
%%sql

SELECT
	cuentas_bancarias.banco AS "Banco",
	COUNT(cuentas_bancarias.id_persona) AS "Cantidad",
	SUM(cuentas_bancarias.saldo) AS "Saldo"
FROM
	personas INNER JOIN cuentas_bancarias
	ON personas.id = cuentas_bancarias.id_persona
GROUP BY
	cuentas_bancarias.banco
ORDER BY
	"Saldo" DESC;

```

A esto lo llamamos nueva_tabla, y seleccionamos de ella la mayor cantidad de usuarios por banco y faccion utilizando un where donde faccion y cantidad de usuarios sean igual a un any dentro del cual se encuentra una tabla que contiene las facciones y su maxima cantidad de usuarios. Finalmente, similar a la consulta 6, pueden existir empates en la cantidad de usuarios maxima por faccion, por lo que seleccionamos una sola utilizando la funcion max en el nombre de los bancos._

```python
%%sql

SELECT  tablita.faccion AS "Facción", 
        MAX(tablita.banco) AS "Banco"
FROM (

    SELECT
        nueva_tabla.faccion AS faccion,
        nueva_tabla.cantidad_usuarios AS cantidad_usuarios,
        nueva_tabla.banco AS banco
    FROM(
        SELECT
            facciones.nombre AS faccion,
            COUNT(cuentas_bancarias.id_persona) AS cantidad_usuarios,
            cuentas_bancarias.banco AS banco
        FROM
            personas INNER JOIN cuentas_bancarias
            ON personas.id = cuentas_bancarias.id_persona
            INNER JOIN persona_pertenece_faccion
            ON personas.id= persona_pertenece_faccion.id_persona
            INNER JOIN facciones
            ON persona_pertenece_faccion.id_faccion = facciones.id
        GROUP BY facciones.nombre, cuentas_bancarias.banco
        ) AS nueva_tabla
    WHERE (nueva_tabla.faccion,nueva_tabla.cantidad_usuarios) = ANY(
        SELECT
            nueva_tabla1.faccion AS "Faccion",
            MAX(nueva_tabla1.cantidad_usuarios) AS "Cantidad Usuarios"
        FROM(
            SELECT
                facciones.nombre AS faccion,
                COUNT(cuentas_bancarias.id_persona) AS cantidad_usuarios,
                cuentas_bancarias.banco AS banco
            FROM
                personas INNER JOIN cuentas_bancarias
                ON personas.id = cuentas_bancarias.id_persona
                INNER JOIN persona_pertenece_faccion
                ON personas.id= persona_pertenece_faccion.id_persona
                INNER JOIN facciones
                ON persona_pertenece_faccion.id_faccion = facciones.id
            GROUP BY facciones.nombre, cuentas_bancarias.banco
            ) AS nueva_tabla1
            GROUP BY nueva_tabla1.faccion
            )) AS tablita
GROUP BY tablita.faccion;
```

### Consulta 10 (6 pts)
**ID, nombre y segundo apellido de las personas traidoras**(Es decir, si la persona pertenece a dos o más facciones.)

Ejemplo:
   
|ID | Nombre | Apellido | 
|-|-------|-|
| 1 | Pedro | Pascal |
| 2 | Bella | Ramsey |

**Solución:** _Realizamos un inner join entre personas y persona_pertenece_faccion, de este selecciamos ID, nombre y apellido (realizando el respectivo corte para que este corresponda al segundo) con la condicion de que la cantidad de relaciones persona_pertenece_faccion.id_faccion de cada persona sea mayor a 1. Luego agrupamos por id, nombre y apellido._

```python
%%sql

SELECT
	personas.id AS "ID",
	personas.nombre AS "Nombre",
	SUBSTRING(personas.apellidos,strpos(personas.apellidos,' ')+1,LENGTH(personas.apellidos)) AS "Apellido"
FROM
	personas INNER JOIN persona_pertenece_faccion
	ON personas.id = persona_pertenece_faccion.id_persona
GROUP BY
	personas.id, personas.nombre,personas.apellidos
HAVING
	COUNT(persona_pertenece_faccion.id_faccion) > 1;
```
